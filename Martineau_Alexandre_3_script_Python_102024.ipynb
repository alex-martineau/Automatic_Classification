```python
import requests
import csv

url = "https://api.edamam.com/api/food-database/v2/parser"
params = {"app_id": "716525e7","app_key": "295f01dd61840cab6c3257b082d59753","ingr": "champagne"}
response = requests.get(url, params=params)

if response.status_code == 200:
    data = response.json()
    products = data.get('hints', [])[:10]
    csv_data = []
    for product in products:
        food = product.get('food', {})
        food_id = food.get('foodId', 'N/A')
        label = food.get('label', 'N/A')
        category = food.get('category', 'N/A')
        food_contents_label = food.get('foodContentsLabel', 'N/A')
        image = food.get('image', 'N/A')
        csv_data.append([food_id, label, category, food_contents_label, image])    
    with open("products_champagne.csv", mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['foodId', 'label', 'category', 'foodContentsLabel', 'image'])
        writer.writerows(csv_data)
    print("Extraction réussie! Les données des produits sont enregistrées dans products_champagne.csv")
else:
    print(f"Erreur {response.status_code}: {response.text}")
```
